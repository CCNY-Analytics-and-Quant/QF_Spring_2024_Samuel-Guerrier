```python
## Once yfinance is installed, you can use the following code to download historical data for your 7 assets.

# Load libraries
import pandas as pd
import yfinance as yf
import datetime as dt

# Create a list of assets (update this with your chosen assets):
my_tickers = ['META', 'AAPL', 'AMZN', 'GOOG', 'NFLX', 'MSFT', 'INTC']

# Assign a start date and end date for the data download, 1 year of data
start_date = dt.datetime.today() - pd.Timedelta(365, "d")
end_date = dt.datetime.today()
```

```python
## Now download the data using the above variables
my_port = yf.download(start=start_date, end=end_date, tickers=my_tickers)
```
