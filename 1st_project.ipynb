```python
import pandas as pd
import yfinance as yf
import datetime as dt

# list of assets (updated with our chosen assets):
my_tickers = ['CRM', 'HPE', 'DELL', 'F', 'TSLA', 'WMT', 'TGT']


# Assign a start date and end date for the data download, 1 year of data
start_date = dt.datetime.today() - pd.Timedelta(3650, "d")
end_date = dt.datetime.today()
```

```python
## Now download the data using the above variables, my_port is short for my portfolio
my_port = yf.download(start=start_date, end=end_date, tickers=my_tickers)
```

```python
## And review the data:
my_port.head()
```

```python
## We may only need one index of the multi-index that was returned (Adj Close)
my_port = my_port['Adj Close']

my_port.head()
```

```python
import pandas as pd

my_port_table = pd.DataFrame(index = my_tickers)
    
```

```python
## Number 1a. for the 1st project
my_port_table
```

```python
## Calculation for 1b. in the 1st project

port_weight = [1/len(my_tickers)] * len(my_tickers)
my_port_table['Portfolio_Weight'] = port_weight

my_port_table

```

```python
## Calculation for 1c. in the 1st project

## Fist calculate daily returns
d_returns = my_port.pct_change()

## Then calculate equally weighted portfolio returns
portfolio_returns = d_returns.dot(port_weight)

## Then calculate standard deviation of portfolio returns for trailing 3 months
rolling_volatility = portfolio_returns.rolling(window=63).std()

## Then annualize the standard deviation
annualized_volatility = rolling_volatility.std() * (252 ** 0.5)  # Assuming 252 trading days in a year

my_port_table['Annualized Volatility'] = annualized_volatility

my_port_table
```
